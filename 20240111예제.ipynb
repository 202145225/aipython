```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'penguins.csv' assuming it is in the current working directory
penguins_data = pd.read_csv('penguins.csv')

# 1 
avg_weight_by_species = penguins_data.groupby(['species', 'sex'])['body_mass_g'].mean().reset_index()

sns.barplot(x='species', y='body_mass_g', hue='sex', data=avg_weight_by_species)
plt.title('    ', fontsize=16)
plt.show()

# 2 
correlation_matrix = penguins_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('penguins  ', fontsize=16)
plt.show()

# 3 
sns.regplot(x='bill_length_mm', y='bill_depth_mm', data=penguins_data)
plt.title('   ', fontsize=16)
plt.show()
```
